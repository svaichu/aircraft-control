#### 2.2.3 Stabilisation via Youla-controller

Consider the following transfer function:

$$
G(s) = \frac{s - 1}{s^3 + s^2 - 4s - 4} = \frac{m(s)}{m(s)}.
$$

Bring the transfer function to a minimum realised transfer function:

$$
G(s) \equiv \begin{bmatrix} A & B \\ C & D \end{bmatrix},
$$

and employ the procedure based on an observer gain $H$ and a state feedback gain $F$, from ACS, Exercise 07, to design a Youla controller with the following coprime factorisation:

$$
\begin{bmatrix} M & U \\ N & V \end{bmatrix} \equiv \begin{bmatrix} A + BF & B - H \\ F & I \end{bmatrix} \begin{bmatrix} I & 0 \\ C + DF & D \end{bmatrix},
$$

where $A + BF$ and $A + HC$ are both stable and satisfy the following modified Bezout's identity:

$$
VM - UN = I.
$$

Design a stabilising controller for $Q = 0$ is given by:

$$
K = -V^{-1}U,
$$

where $U(s), V(s) \in RH_\infty$. Check internal stability for your controller. Is it fulfilled for the given $G(s)$? Document all your steps by the corresponding formula and numerical matrices and come to a conclusion.
