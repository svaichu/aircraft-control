### 2.1 System Analysis  
2.1.1 Zero
Calculte system zero from state space representation without using Control System Toolbox of MATLAB.

```matlab
% Create a random A, B, C, D 
% input size p and output size m
% from zeros and poles
p = 4; % input size
m = p; % output size
s = tf('s');

sys = rss(5, m, p);

A = sys.A; B = sys.B; C = sys.C; D = sys.D;
```

```matlab
sys.D
```

```matlab
% Calculate system zero without using Control System Toolbox
n = size(A,1);
M = [s*eye(n)-A, -B; C, D];
zero_eq = simplify(det(M));
zeros = double(solve(zero_eq, s));
disp('System Zeros are:');
disp(zeros);
```

```matlab
% Define state-space matrices
A = [5 1; -2 -3];
B = [0; 1];
C = [1 0];
D = 0;

% Form the generalized eigenvalue problem
E = [A B; C D];
F = [eye(size(A)) zeros(size(B)); zeros(size(C)) zeros(size(D))];

% Solve for the zeros
zeros_ss = eig(E, F);

% Remove infinite or invalid zeros
zeros_ss = zeros_ss(~isinf(zeros_ss) & ~isnan(zeros_ss));

% Display the zeros
% disp('Zeros of the state-space system:')
zeros_ss
```
